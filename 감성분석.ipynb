```python
from transformers import TextClassificationPipeline, BertForSequenceClassification, AutoTokenizer
model_name = 'smilegate-ai/kor_unsmile'
model = BertForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
pipe = TextClassificationPipeline(
        model = model,
        tokenizer = tokenizer,
        device = 0,   # cpu: -1, gpu: gpu number
        return_all_scores = True,
        function_to_apply = 'sigmoid'
    )
#print(data_1['제목'][33])

def max_index(b) :
  x = 0
  for i in range(1, len(b)) :
    if b[i] > b[x] :
      x = i
  return x
 
a = ['여성/가족', '남성', '성소수자', '인종/국적','연령','지역','종교','기타 혐오','악플/욕설', 'clean']
b = []
c = []
for i in range(len(data_1)) :
  #print(i)
  for result in pipe(data_1['제목'][i])[0]:
    #if result['label'] == '기타 혐오' or result['label'] == '악플/욕설'
    #print(result)
    b.append(result['score'])
  c.append(a[max_index(b)])  
  b = []

d = []
for i in a :
  d.append(c.count(i)/len(c)) # 최대로 많이 나오는 값 세서 붙이기
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette('hls',len(a))
plt.figure(figsize=(13, 8))
plt.xticks(fontsize=15)
plt.yticks(np.arange(0, 1, 0.1), ('0%', '10%', '20%', '30%', '40%', '50%', '60%', '70%', '80%'), fontsize=15)
plt.title('2021년 11월 기사 제목 별 감성 분석', fontsize=20)
plt.bar(a, d, color = colors)
```
